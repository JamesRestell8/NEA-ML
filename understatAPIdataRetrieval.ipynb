```python
import pandas as pd
import requests
from requests.exceptions import ConnectionError
import time
from understat import Understat
import asyncio
import json
import aiohttp
import nest_asyncio
from pandas.io.json import json_normalize

nest_asyncio.apply()
```

```python
async def main():
    async with aiohttp.ClientSession() as session:
        understat = Understat(session)
        player = await understat.get_league_players(
        "epl", 2021
    )
    return (json.dumps(player))

loop = asyncio.get_event_loop()
allPlayerStats = loop.run_until_complete(main())
print(allPlayerStats)
allPlayerStats = pd.read_json(allPlayerStats)
```

```python
allPlayerStats
```

```python
allPlayerStats.drop(allPlayerStats[allPlayerStats.time < 300].index, inplace=True)
allPlayerStats.drop(allPlayerStats[allPlayerStats.team_title == "Norwich"].index, inplace=True)
allPlayerStats.drop(allPlayerStats[allPlayerStats.team_title == "Watford"].index, inplace=True)
allPlayerStats.drop(allPlayerStats[allPlayerStats.team_title == "Burnley"].index, inplace=True)
```

```python
del allPlayerStats['games']
del allPlayerStats['assists']
del allPlayerStats['red_cards']
del allPlayerStats['position']
del allPlayerStats['team_title']
del allPlayerStats['npg']
del allPlayerStats['goals']
```

We will now merge these dataframes with the FPL dataframes in the other notebook.

```python
async def main():
    async with aiohttp.ClientSession() as session:
        understat = Understat(session)
        player = await understat.get_league_players(
        "epl", 2019
    )
    return (json.dumps(player))

loop = asyncio.get_event_loop()
allPlayerStats = loop.run_until_complete(main())
print(allPlayerStats)
allPlayerStats = pd.read_json(allPlayerStats)

allPlayerStats.drop(allPlayerStats[allPlayerStats.time < 300].index, inplace=True)

del allPlayerStats['games']
del allPlayerStats['assists']
del allPlayerStats['red_cards']
del allPlayerStats['position']
del allPlayerStats['team_title']
del allPlayerStats['npg']
del allPlayerStats['goals']

allPlayerStats
```
